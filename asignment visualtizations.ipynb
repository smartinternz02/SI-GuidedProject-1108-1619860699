```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel("x")
plt.ylabel("y")
plt.title("title")
```

```python
plt.scatter(x,y,color="purple")
plt.xlabel("x")
plt.ylabel("y")
plt.title("title")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,4)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.scatter(x,y,color ="purple")
plt.subplot(2,2,3)
plt.scatter(x,y,color ="pink")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.2,0.2])
axes2 = fig.add_axes([0.1,0.1,0.5,0.5])
axes.plot(x,y,'r')
axes2.scatter(x,y,color='green')
```

```python
plt.bar(x,y,color = "red")
```

```python
plt.plot(x,y)
scale_factor = 0.2
xmin,xmax = plt.xlim()
ymin,ymax = plt.ylim()
plt.xlim(xmin*scale_factor,xmax*scale_factor)
plt.ylim(ymin*scale_factor,ymax*scale_factor)
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x = "total_bill",y = "tip",data = tips)
```

```python
plt.scatter(tips["total_bill"],tips["tip"])
```

```python
sns.lineplot(x="size",y="tip",data = tips)
```

```python
#prices and tips are directly proportional
```

```python
sns.distplot(tips["tip"])
```

```python
#most of tips given are by the people who ordered between 2 and 4 $/-\
```

```python
sns.barplot(x = "time",y="tip",data = tips)
```

```python
#tips are high during dinner
```

```python
sns.boxplot(x="size", y="tip",data = tips)
```

```python
#confined to number of the player's visited and the tip's given.
```

```python
sns.catplot(x="smoker",y="tip",data = tips)
```

```python
#most tips are given by non smokers.
```

```python
sns.lineplot(x="smoker",y="tip",data = tips)
```

```python
sns.pairplot(tips)
```

```python
sns.heatmap(tips.corr(),annot = True)
```

```python
#gives values from -1 to 1
#>0.5,0.6,0.8 -highly positively correlated(directly proportional)
#<0.5-partially correlated
#>-0.5 -partially 
#<-0.5 highly negatively correlated
```
